```python
# Creation of set
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung'}
print(brands)
```

```python
# Duplicates not allowed
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung', 'apple'}
print(brands)
```

```python
# The values true and 1 are considered the same value in sets
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung', True, 1, 0, False, 2}
print(brands)
```

```python
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung'}
print('dell' in brands)
print('hp' in brands)
```

```python
# Once a set is created you cannot replace its items, but you can add new items
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung'}
brands.add('hp')
print(brands)
```

```python
# Remove items
brands = {'apple', 'nike', 'adidas', 'dell', 'samsung'}
brands.remove('dell')
print(brands)
```

```python
# Set operations
A = {5, 6, 7 ,8}
B = {1, 2, 5, 6}

print('Union: ', A | B)
print('Intersection: ', A & B)
print('Difference: ', B -A)
print('Symmetric Difference: ', A ^ B)
```

```python
# Frozenset
brands = frozenset(['apple', 'nike', 'adidas', 'dell', 'samsung'])
print(brands)
print(type(brands))
```

```python
# Adding or Removal of elements are not allowed in Frozenset
brands.add('hp')
brands.remove('samsung')
print(brands)
```
